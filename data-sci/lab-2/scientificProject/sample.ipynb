### Вбудовані способи вибору

У бібліотеці Pandas існує кілька способів вибору даних із DataFrame. Кожен із них має свої особливості.

1. **Доступ до стовпця, як до атрибута об'єкта**:
   - Можна звернутись до стовпця ніби до атрибута, що доволі красиво виглядає.
   - Цей спосіб працює лише для тих стовпців, назви яких є валідними іменами змінних у Python, тобто не можна пробілів, спеціальних символів та повинні починатись з літери.
2. **Доступ до стовпця по ключу словника**:
   - Також можна звертатись як до словника, що дозволяє обирати всі імена, оскільки вони передаються як рядок.
   - Цей спосіб мені більше подобається, оскільки працює універсально для будь-якого імені.
3. **Доступ до елемента стовпця по індексу**:
   - Використовуючи індекси, можна вибирати окремі елементи стовпця за допомогою нотації масиву.
   - Це доволі просто виглядає, але є спеціалізовані методи які працюють ефективніше

```python
# Доступ до стовпця MouseID як атрибуту об'єкта
data.MouseID
```

```python
# Доступ до стовпця MouseID по ключу словника
data['MouseID']
```

```python
# Доступ до елемента стовпця MouseID по індексу
data['MouseID'][3]
```

### Індексно‐орієнтований вибір та мітко‐орієнтований вибір

В Pandas є два головні методи для вибору даних за індексом: `iloc` і `loc`. Вони допомагають обирати дані з DataFrame на основі індексів або міток.

1. **`iloc` — вибір за позицією (індексами)**:
   - Цей метод використовується, коли потрібно обрати рядки або стовпці за їх порядковим номером.
   - Наприклад, можна взяти перший рядок або стовпець. Це корисно, коли нам важливо знати позицію даних, а не їхню мітку.
   - Мені здається, це найпростіший спосіб, при праці з великими таблицями й треба швидко дістатись до конкретних рядків або стовпців.

2. **`loc` — вибір за міткою (назвами рядків або стовпців)**:
   - Цей метод гарний тим, що можна вибирати дані за назвами стовпців або індексів. Тобто, замість використання числових індексів, можна писати назви стовпців.
   - Я використовую цей метод, коли хочу працювати з конкретними назвами стовпців або хочу обрати кілька рядків за їхніми мітками.
   - Це зручно, коли не відомо точні позиції, але знаємо назви колонок або індекси рядків.

```python
# Вибираємо перший рядок
data.iloc[0]
```

```python
# Вибираємо другий стовпець
data.iloc[:, 1]
```

```python
# Вибираємо перші три рядки другого стовпця
data.iloc[:3, 1]
```

```python
# Вибираємо елемент за міткою: рядок 0 і стовпець 'MouseID'
data.loc[0, 'MouseID']
```

```python
# Вибираємо стовпці 'Genotype', 'Treatment', 'Behavior' для перших п'яти рядків
data.loc[:5, ['Genotype', 'Treatment', 'Behavior']]
```

### Умовний вибір

У Pandas умовний вибір — це можливість фільтрувати дані за певними умовами, що робить роботу з DataFrame гнучкою і ефективною. Ми можемо швидко виділяти потрібні рядки або стовпці, перевіряючи їх на відповідність умовам, що робить цей інструмент неймовірно корисним для аналізу даних.

#### Проста умова

Проста умова — це коли ми порівнюємо значення в стовпці з певним числом чи іншим значенням. Наприклад, якщо нам треба вибрати всі записи, де значення якогось показника більше або менше певної величини, це зробити дуже просто. 

Ось приклад: вибираємо всі записи, де рівень білка **BDNF_N** більший за середнє значення по цьому стовпцю.

```python
# Обчислюємо середнє значення для стовпця BDNF_N
mean_bdnf = data.BDNF_N.mean()
# Вибираємо всі рядки, де значення BDNF_N більше за середнє
data_above_mean_bdnf = data.loc[data.BDNF_N > mean_bdnf]
print(mean_bdnf)
data_above_mean_bdnf[['BDNF_N', 'MouseID']]
```

#### Складна умова

Складні умови дозволяють комбінувати кілька простих умов за допомогою логічних операторів `&` (AND) та `|` (OR). Це зручно, коли нам треба накласти кілька фільтрів одночасно. Наприклад, можна вибрати всі рядки, де рівень білка **BDNF_N** менший за середнє і миша має генотип **Control**.



```python
# Вибираємо рядки з комбінованими умовами: BDNF_N < середнього та Genotype == 'Control'
data_filtered = data.loc[(data.BDNF_N < mean_bdnf) & (data.Genotype == 'Control')]
data_filtered
```

#### Умовні селектори

`isin()` — це класний метод, який дозволяє вибирати рядки, де значення стовпця належить певному списку значень. Це дуже зручно, коли ти хочеш відфільтрувати дані за кількома конкретними значеннями. Наприклад, якщо ми хочемо вибрати всі записи, де мишам давали лікування **Memantine** або **Saline**.

```python
# Вибираємо рядки, де Treatment дорівнює 'Memantine' або 'Saline'
data_selected = data.loc[data.Treatment.isin(['Memantine', 'Saline'])]
data_selected
```

#### Series як умова

У Pandas, коли ми робимо умовний вибір даних, насправді ми створюємо **Series**, яка містить булеві значення (`True` або `False`). Кожен елемент цього Series відповідає рядку в DataFrame і показує, чи виконується умова для конкретного рядка.

Цей булевий Series діє як фільтр, який дозволяє Pandas обрати тільки ті рядки, де значення є `True`. Умови можуть бути різними: порівняння чисел, рядків або навіть кілька умов, об'єднаних логічними операторами. Це робить роботу з даними дуже ефективною, оскільки можна швидко відсіяти непотрібну інформацію, залишивши лише те, що відповідає нашим вимогам.

```python
data.Treatment.isin(['Memantine', 'Saline'])
```

### Швидкодія та реалізація

Pandas побудована на основі бібліотеки **NumPy**, яка, в свою чергу, використовує мову програмування **C** для виконання багатьох операцій. Завдяки цьому більшість операцій із даними в Pandas працюють дуже швидко, оскільки вони фактично виконуються на низькому рівні, безпосередньо в пам'яті, без значних накладних витрат Python.

Pandas оптимізує багато операцій завдяки використанню **векторизації**. Це означає, що операції застосовуються до всього масиву даних одразу, а не ітеруються по кожному елементу окремо, що було б значно повільніше. Векторизація дозволяє уникати циклів Python, замінюючи їх на операції, реалізовані на рівні C.

Також Pandas використовує **булеві маски** та індексацію, що дозволяє дуже швидко виконувати умовні вибори даних, порівняння і фільтрацію. Замість того, щоб вручну перевіряти кожен рядок, ці операції виконуються одночасно для всіх рядків, завдяки чому обробка великих обсягів даних стає більш ефективною.

Завдяки цим оптимізаціям, Pandas є дуже швидким інструментом для роботи з великими наборами даних, особливо коли потрібна фільтрація, агрегація чи обчислення середніх або інших статистичних параметрів.
