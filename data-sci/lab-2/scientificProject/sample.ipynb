```python
# Обчислюємо середнє значення для стовпця BDNF_N
mean_bdnf = data.BDNF_N.mean()
# Вибираємо всі рядки, де значення BDNF_N більше за середнє
data_above_mean_bdnf = data.loc[data.BDNF_N > mean_bdnf]
print(mean_bdnf)
data_above_mean_bdnf[['BDNF_N', 'MouseID']]
```

```python
# Вибираємо рядки з комбінованими умовами: BDNF_N < середнього та Genotype == 'Control'
data_filtered = data.loc[(data.BDNF_N < mean_bdnf) & (data.Genotype == 'Control')]
data_filtered
```

```python
# Вибираємо рядки, де Treatment дорівнює 'Memantine' або 'Saline'
data_selected = data.loc[data.Treatment.isin(['Memantine', 'Saline'])]
data_selected
```
